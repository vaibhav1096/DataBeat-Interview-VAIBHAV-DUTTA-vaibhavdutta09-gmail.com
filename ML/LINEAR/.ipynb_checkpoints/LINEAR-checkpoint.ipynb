```python
m=df['quality'].values.size
m
```

```python
df_zscore = (df - df.mean()) / df.std()
df_zscore
```

```python
df_zscore['ones']=1
df_zscore
```

```python
y=df['quality']
X=df_zscore.drop(['quality'],axis=1)
X = np.matrix(X.values)
y = np.matrix(y.values).T
```

```python
theta = np.matrix(np.zeros(shape=[1,X.shape[1]]))
```

```python
y.shape,X.shape,theta.shape
```

```python
def cost_function(X,y,theta):
    y_pred=np.dot(X,theta.T)
    sqrd_error=np.power((y_pred-y),2)
    cost=(1/(2*m))*(np.sum(sqrd_error))
    return cost

```

```python
def gradient_descent(X,y,theta,alpha,iterations):
    costs=[]
    theta=theta.T
    for i in range (iterations):
        y_pred=np.dot(X,theta)
        der=np.dot(X.transpose(),(y_pred-y))/m
        theta=theta-alpha*der
        costs.append(cost_function(X,y,theta.T))
    return theta,costs

```

```python
theta_final,cost=gradient_descent(X,y,theta,alpha=0.00068,iterations=10000)
theta_final,cost[-1]
```

```python
for i in range(0, 11):
    print(df_zscore.columns[i], theta_final[i])
```
